```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv

spreadsheet = pd.read_csv("C:/Users/eaindray/Downloads/healthy_lifestyle_city_2021 (1).csv")
spreadsheet.describe()
```

```python
spreadsheet.dropna(inplace=True)
spreadsheet.head()
```

```python
spreadsheet.info()
```

```python
import seaborn as sns
sns.pairplot(spreadsheet)
```

```python
corr=spreadsheet.corr(numeric_only = True)
corr
```

```python

numeric_columns = spreadsheet.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
corr_matrix = numeric_columns.corr()

# Plot the correlation heatmap
sns.heatmap(corr_matrix, vmin=-1, vmax=1, center=0, annot=True)

# Display the plot
plt.show()

```
